3. visualizzare le province della Lombardia su una mappa di contextily

```python
Lombardia=regioni[regioni.DEN_REG== 'Lombardia']
Lombardia
```

```python
province_lombardia=Province[Province.within(Lombardia.geometry.squeeze())]
province_lombardia
```

```python
ax=province_lombardia.to_crs(epsg=3857).plot(facecolor='none',edgecolor='k')
ctx.add_basemap(ax)
```

```python
Utente=input('Inserire una provincia della lombardia: ')
lst=list(province_lombardia['DEN_UTS'])
if Utente not in lst:
  print('provincia non in lombardia')
else:
 prov=province_lombardia[province_lombardia['DEN_UTS']==Utente].geometry.squeeze()
 prov2=province_lombardia[province_lombardia['DEN_UTS']==Utente]
 ax=province_lombardia[province_lombardia.touches(prov)].to_crs(epsg=3857).plot(edgecolor='k',alpha=0.5,figsize=(15,10))
 prov2.to_crs(epsg=3857).plot(ax=ax,edgecolor='r',alpha=0.5)
 ctx.add_basemap(ax)
```

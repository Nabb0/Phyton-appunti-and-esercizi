Vediamo come è possibile unire due dataframe che presentano la stessa struttura:

```python
df_multe_complessivo = df_multe.append(df_multe1) #unisce 2 dataframe con colonnne uguali gia presenti ad esempio multe del 20/21
df_multe_complessivo
```

```python
df_multe_complessivo = df_multe.append(df_multe1).reset_index() #crea il nuovo indice con tutti i numeri
df_multe_complessivo
```

```python
df_multe_complessivo = df_multe.append(df_multe1, ignore_index=True) #mette il nuovo indece e elimina quello vecchio 
df_multe_complessivo
```

```python
df_multe_complessivo = pd.concat([df_multe, df_multe1]) #contatena 
df_multe_complessivo
```

Vediamo ora come unire le informazioni presenti in dataframe che hanno una struttura diversa tranne che per un campo (o più campi) che contengono informazioni che permettono di collegare un record (una riga) di un dataframe con uno o più record dell'altro dataframe:

<img src="https://miro.medium.com/max/1838/1*ZCpo3gXuXI4KFhKivEt2ZA.png" width=800>

Nell'esempio il campo "id" permette di mettere in relazione le righe del primo dataframe con le righe del secondo dataframe. Prestare attenzione al fatto che il legame non è causato tanto dal nome del campo (che è lo stesso per rendere più leggibile la relazione) quanto dal valore contenuto in quel campo. L'id 2 indica Jenny in un dataframe e una donna di 31 anni nel secondo dataframe: siccome l'id è lo stesso, vuol dire che Jenny ha 31 anni.

Per fare questa operazione di join utilizzeremo il metodo di Pandas merge() che effettua una join (una associazione, una connessione, un collegamento) tra i record dei due dataframe dove il campo in comune contiene lo stesso valore.

Ci sono diversi tipi di join, ognuna che effettua una associazione dal risultato diverso. Le join principali sono la left join, la right join, la outer join e la inner join.

<img src="https://i1.wp.com/datascienceparichay.com/wp-content/uploads/2021/01/Left-Join-diagram.png?w=800&ssl=1">

```python
df_proprietari_multe = pd.merge(df_prop, df_multe_complessivo, how="left", on=["targa"]) #nat not a time ,matcha i dati col database di sinistra, se nella tabella di destra ci sono dati in piu non vengono presi
df_proprietari_multe # qui abbiamo i tutti i propretari con le loro multe
```

<img src="https://i2.wp.com/datascienceparichay.com/wp-content/uploads/2021/01/Right-Join-diagram.png?w=800&ssl=1">

```python
df_proprietari_multe = pd.merge(df_prop, df_multe_complessivo, how="right", on=["targa"])#, se nella tabella di sinistra ci sono dati in piu non vengono presi,matcha i dati col database di destra,
df_proprietari_multe #qui abbiamo tutte le multe con i loro propretari
```

<img src="https://i1.wp.com/datascienceparichay.com/wp-content/uploads/2021/01/Outer-Join-diagram.png?w=800&ssl=1" >

```python
df_proprietari_multe = pd.merge(df_prop, df_multe_complessivo, how="outer", on=["targa"]) #prendo tutte e due le tabelle dove riesco a unire unisco dove non riesco li riporto con dato mancante
df_proprietari_multe
```

<img src="https://i0.wp.com/datascienceparichay.com/wp-content/uploads/2021/01/Inner-Join-diagram.png?w=800&ssl=1">

```python
df_proprietari_multe = pd.merge(df_prop, df_multe_complessivo, how="inner", on=["targa"]) #esclude i tutti dati senza una corrispondenza
df_proprietari_multe
```

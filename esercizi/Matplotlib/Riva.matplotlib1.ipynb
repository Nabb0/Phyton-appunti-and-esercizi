# ***Esercizio 1***

a. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python


data = {
  'last_name': ['Boyer', 'Copeland', 'Daniel','Serrano','Tarrell','Vargas'],
  'order_numbers': [185, 50, 196,58,24,28]
 
}

df = pd.DataFrame(data, index=[0,1,2,3,4,5])
df
```

b. visualizzi un grafico a barre verticali per rappresentare i dati. Il risultato deve essere questo:

```python

x=list(df['last_name'])

y=list(df['order_numbers'])

fig = plt.figure(figsize=(7,8))
ax = plt.axes()
plt.rcParams.update({'font.size':9}) #grandezza delle scritte
ax.bar(x, y) #ascisse,ordinate=cordinare del piano cartesiano

```

c. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python

fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'],df['order_numbers'],color='#FFA07A')
ax.set_xlabel('commessi')
ax.set_ylabel('ordini')
fig.suptitle('Numeri di ordini dai vari commessi')
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df.sort_values('order_numbers',ascending=False)['last_name'],df['order_numbers'].sort_values(ascending=False),color='#5F9EA0')
```

e. visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

```python

fig = plt.figure()
ax = plt.axes()
ax.barh(df['last_name'],df['order_numbers'],color='#FFA07A')
ax.set_xlabel('commessi')
ax.set_ylabel('ordini')
fig.suptitle('Numeri di ordini dai vari commessi')
```

f. crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
tot=185+ 50+ 196+58+24+28

data2 = {
  'last_name': ['Copeland', 'Serrano', 'Boyer','Daniel','Vargas','Tarrell'],

  'orders_perc': [round((50*100)/tot,2),round((58*100)/tot,2),round((185*100)/tot,2),round((196*100)/tot,2),round((28*100)/tot,2),round((24*100)/tot,2)]
 
}

df1 = pd.DataFrame(data2, index=[0,1,2,3,4,5])
df1
```

g. visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python



plt.rcParams.update({'font.size': 15})    # Aumenta la grandezza del testo (Default: 10)

fig = plt.figure(figsize=[7,7])
ax = plt.axes()
ax.pie(df1['orders_perc'], labels=df1['last_name'], autopct='%1.1f%%')



plt.show()
```

h. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python

plt.rcParams.update({'font.size': 15})    # Aumenta la grandezza del testo (Default: 10)

fig = plt.figure(figsize=[7,7])
ax = plt.axes()
cols = ['#00FFFF','#A52A2A','#5F9EA0','#7FFF00','#B8860B','#A9A9A9']
ax.pie(df1['orders_perc'], labels=df1['last_name'], autopct='%1.1f%%',colors=cols)


fig.suptitle('Report Vendite', color='k')



plt.show()
```

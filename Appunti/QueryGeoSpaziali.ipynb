```python
Reg_prov= geopandas.sjoin(Province2,Regioni2,op='within',how='left') # fa una join spaziale tra due dataframe 
Nprovince= Reg_prov.groupby('DEN_REG')[['DEN_UTS']].count().reset_index() # fa un gruopby per sapere le province per ogni regione 
regioniprovince=pd.merge(Regioni2,Nprovince,how='left',on='DEN_REG') # merge per unire i dati ottenutu
regioniprovince.plot(legend=True,column='DEN_UTS') # grafico dell italia con il numero delle  province 
```

colorare l' italia in base i Comuni per ogni provincia

```python
#trova per ogni comune la provincia in  cui si trova quel comune
#effetuando una join spaziale: prendendo la geometria del comune
#e cerca in quale geometria della provincia si trova.
#ristituisce un geosataframe 
com_prov=geopandas.sjoin(comuni2,Province2,op='within',how='left')

#raggruppa i comuni di ogni provincia e li conta.
#restituisce un dataframe
Ncomuni= com_prov.groupby('DEN_UTS')['COMUNE'].count().reset_index()

#unisce il geodatframe delle province con il datframe che contiene il numero di 
# ogni provincia.
#restituisce un geodataframe che e uguale a Province2 ma con una colonna in piu
# che è il numero di comuni per ogni provincia
ComuniProvince=pd.merge(Province2,Ncomuni,how='left',on='DEN_UTS')

#Visualiazza le province colorate in base al numero di comuni(column='COMUNE')
ComuniProvince.plot(legend=True,column='COMUNE')
```
